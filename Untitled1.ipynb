```python
# import necessary libraries
import pandas as pd
import numpy as np
import json
from datetime import datetime
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import shap
from textblob import TextBlob
from feature_engine.outliers import Winsorizer
from tensorflow import keras
```

```python
# Data Loading and Preparation
df = pd.read_excel(r"/content/drive/MyDrive/Homestays_Data.xlsx")
imp_features = ['amenities', 'host_since', 'last_review']
df1 = df.loc[:, imp_features]
imp_features1 = ['latitude', 'longitude', 'neighbourhood', 'log_price']
df2 = df.loc[:, imp_features1]
imp_features2 = ['room_type', 'city', 'property_type']
df5 = df.loc[:, imp_features2]
imp_features3 = ['log_price', 'accommodates', 'bathrooms', 'review_scores_rating', 'bedrooms', 'beds']
df6 = df.loc[:, imp_features3]
```

```python
# Feature Engineering
today = datetime.today()
df1['host_since'] = pd.to_datetime(df1['host_since'])
df1['Host_Tenure'] = (today.year - df1['host_since'].dt.year) + (today.month - df1['host_since'].dt.month) / 12
df1['Host_Tenure'] = df1['Host_Tenure'].clip(lower=0)


def count_amenities(amenities_str):
    try:
        amenities_list = json.loads(amenities_str.strip()[1:-1])
    except json.JSONDecodeError:
        amenities_list = amenities_str.strip()[1:-1].split(',')
    return len(amenities_list)

df1['Amenities_Count'] = df1['amenities'].apply(count_amenities)
df1['last_review'] = pd.to_datetime(df1['last_review'])
df1['Days_Since_Last_Review'] = (today - df1['last_review']).dt.days
```

```python
# Geospatial analysis:

neighborhood_prices = {}
for index, row in df2.iterrows():
    neighborhood = row['neighbourhood']
    price = df2['log_price'].iloc[index]
    if neighborhood not in neighborhood_prices:
        neighborhood_prices[neighborhood] = []
    neighborhood_prices[neighborhood].append(price)

for neighborhood, prices in neighborhood_prices.items():
    average_price = sum(prices) / len(prices)
    print(f"Neighborhood: {neighborhood}, Average Price: ${average_price:.2f}")
```

```python

# Data preprocessing(data cleaning):-
log_price_column = 'log_price'
X = new_df.drop(log_price_column, axis=1)
Y = new_df[log_price_column]

X.isnull().sum() # missing values are present
X.plot(kind = 'box', subplots = True, sharey = False, figsize = (20, 8)) # here we see huge outliers are present here.

```

```python
# Data Preprocessing:-
num_pipeline = Pipeline(steps=[('impute', SimpleImputer(strategy='mean'))])
preprocessor = ColumnTransformer(transformers=[('num', num_pipeline, X.columns)])
imputation = preprocessor.fit(X)
clean = pd.DataFrame(imputation.transform(X), columns=X.columns)

winsor = Winsorizer(capping_method='gaussian', tail='both', fold=0.20, variables=list(clean.columns))
clean1 = winsor.fit(clean)
cleandata = pd.DataFrame(clean1.transform(clean), columns=clean.columns)

scale_pipeline = Pipeline([('scale', MinMaxScaler())])
columntransfer = ColumnTransformer([('scale', scale_pipeline, X.columns)])
scale = columntransfer.fit(cleandata)
scaled_data = pd.DataFrame(scale.transform(cleandata), columns=cleandata.columns)
```

```python
scaled_data
```

```python
# 6. Model Building
# Split data into training and testing sets (optional but recommended)
X_train, X_test, y_train, y_test = train_test_split(scaled_data, Y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
mse1 = mean_squared_error(y_test, y_predicted)
print("Linear Regression Mean Squared Error:", mse1)
```

```python
# 7. Model Optimization and Validation

param_grid = {'fit_intercept': [True, False]}
kfold = KFold(n_splits=5, shuffle=True)
grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=kfold)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
best_mse = grid_search.best_score_ * -1
print("Best Mean Squared Error (GridSearchCV):", best_mse)
y_predicted = best_model.predict(X_test)
final_mse = mean_squared_error(y_test, y_predicted)
print("Final Mean Squared Error on Test Set:", final_mse)
```

```python
# 8. Alternative Models
model1 = RandomForestRegressor(n_estimators=100, random_state=42)
model1.fit(X_train, y_train)
y_predicted = model1.predict(X_test)
mse2 = mean_squared_error(y_test, y_predicted)
print("Random Forest Mean Squared Error:", mse2)

model2 = XGBRegressor(objective='reg:squarederror', max_depth=5, n_estimators=100)
model2.fit(X_train, y_train)
y_predicted = model2.predict(X_test)
mse3 = mean_squared_error(y_test, y_predicted)
print("XGBoost Mean Squared Error:", mse3)


# Define the ANN model:-
model_ann = keras.Sequential([
    keras.layers.Dense(units=32, activation="relu", input_shape=(X_train.shape[1],)),
    keras.layers.Dense(units=16, activation="relu"),
    keras.layers.Dense(units=1)
])
# Compile the ANN model
model_ann.compile(optimizer="adam", loss="mse", metrics=['accuracy'])
# Train the ANN model
model_ann.fit(X_train, y_train, epochs=15, batch_size=32)
# Predict using the ANN model
y_predicted_ann = model_ann.predict(X_test)
# Evaluate the ANN model
mse_ann = mean_squared_error(y_test, y_predicted_ann)
print("ANN Mean Squared Error:", mse_ann)


```

```python
def analyze_feature_importance(model, X_test, y_test, feature_names):
    results = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
    for name, importance in zip(feature_names, results.importances_mean):
        print(f"Feature: {name}, Importance (Tree-Based): {importance:.4f}")

analyze_feature_importance(RandomForestRegressor().fit(X_train, y_train), X_test, y_test, X_test.columns)
```


1. Accuracy Metrics:

RMSE (Root Mean Squared Error): This measures the average difference between model's predictions and the actual values. Lower RMSE is better, indicating the model's predictions are generally close to the real values.
R-squared: This tells us how well our model explains the data's variation. A value closer to 1 signifies a better fit, where 1 means perfect prediction. However, a very high R-squared might indicate overfitting (the model memorizes the training data too well and doesn't generalize well to unseen data).

2. Digging Deeper: Residual Analysis

Residuals: These are the differences between the actual values and the model's predictions.
Residuals vs. True Values Plot: This plot shows how the residuals vary with the actual values. We want no clear patterns here. If there's a consistent trend (e.g., residuals increase for higher true values), it suggests the model might be biased towards underestimating or overestimating for certain value ranges.
Distribution of Residuals Plot: This is a histogram showing how frequently different residual values occur. Ideally, the residuals should be centered around zero and spread out evenly, suggesting the model's errors are random and not skewed in any particular direction.
